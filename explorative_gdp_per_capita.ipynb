### Exploring GDP per Capita Based on Purchasing Power Parity (PPP) Across Countries

The dataset used in this analysis was obtained from Our World in Data. It provides valuable information on the Gross Domestic Product (GDP) of various countries, measured using Purchasing Power Parity (PPP). GDP per capita is a widely utilized economic indicator that reflects the average wealth or income of individuals within a country. By considering PPP, this dataset allows for a more accurate comparison of economic well-being across different nations, as it takes into account variations in the cost of living and currency exchange rates.

The dataset serves as a valuable resource for analyzing and comprehending the economic landscape across countries. It enables researchers and analysts to study the disparities and similarities in economic performance, identify trends, and draw meaningful insights. Accessible at source link, this dataset offers a comprehensive perspective on GDP per capita based on PPP, facilitating a deeper understanding of the global economic dynamics.

Source Link: https://ourworldindata.org/grapher/gdp-per-capita-worldbank

```python
#Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the CSV file into a pandas DataFrame
df = pd.read_csv('gdp-per-capita-worldbank.csv', encoding='latin1')
```

```python
df.head(20)
```

```python
import pandas as pd

# pivot df to convert years as columns
df = df.pivot_table(index=['Entity', 'Code'], columns='Year', values='GDP per capita, PPP (constant 2017 international $)').reset_index()

# Rename the columns
df = df.rename(columns={'Entity': 'Country ', 'Code': 'Country Code'})

# Remove the column name from the index column
df = df.rename_axis(columns=None)

# Print the pivoted dataframe
print(df)
```

```python
df.head(10)
```

```python
# check how many row are in Country Column
df.shape[0]
```

```python
# Dupicate in Country column
df['Country '].duplicated().any()
```

```python
#Get the columns name
column_names = df.columns.tolist()
print(column_names)
```

```python
#For getting Columns having at least 1 null value (column names)
df.columns[df.isnull().any()]
```

```python
# For getting Columns with count, with having at least 1 null value.
df[df.columns[df.isnull().any()]].isnull().sum()
```

```python
# For getting percentage of the null count.
df[df.columns[df.isnull().any()]].isnull().sum() * 100 / df.shape[0]
```

```python
# For getting percentage of the null count and draw graph to visulize.

null_percentages = df[df.columns[df.isnull().any()]].isnull().sum() * 100 / df.shape[0]

# Create a bar plot
null_percentages.plot(kind='bar', figsize=(10, 6))
plt.title('Null Percentage by Column')
plt.xlabel('Columns')
plt.ylabel('Null Percentage')
plt.xticks(rotation=45)
plt.show()
```

### Analysis of Missing Data: Null Percentages by Yea
